### Scrapping 

```python
prod = soup.select("._2WkVRV")
```

```python
prod
```

```python
product_name = []
```

```python
for item in prod:
    product_name.append(item.text)
```

```python
product_name
```

```python
prize = soup.select("._3bPFwb ._30jeq3")
```

```python
prize
```

```python
cost = []
```

```python
for item in prize:
    cost.append(item.text)
```

```python
cost
```

### Creating Dataframe

```python
import pandas as pd
```

```python
progo = pd.DataFrame(product_name, columns=['Product']) 
```

```python
progo
```

```python
progo['Cost'] = cost
```
